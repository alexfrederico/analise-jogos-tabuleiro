# Pré processando os dados

```python
# Separando os mecânicas de cada jogo
dstack = df['mecanicas'].str.split(',', expand=True).stack().str.strip().to_frame()
```

```python
dstack
```

```python
# Transformando as mecânicas em colunas
df_mecanicas = pd.get_dummies(dstack,prefix='',prefix_sep='')
```

```python
df_mecanicas
```

```python
# Agrupando as mecânicas por jogo
df_mecanicas = df_mecanicas.groupby(level=0).sum()
```

```python
df_mecanicas
```

```python
# juntando o dataframe original com o dataframe de mecânicas
df_new = pd.merge(df, df_mecanicas, left_index=True, right_index=True)
```

```python
df_new
```

```python
# Repetindo o processo anterior para a coluna de dominios
dstack = df['dominio'].str.split(',', expand=True).stack().str.strip().to_frame()
df_dominio = pd.get_dummies(dstack,prefix='',prefix_sep='')
df_dominio = df_dominio.groupby(level=0).sum()
df_new = pd.merge(df_new, df_dominio, left_index=True, right_index=True)
```

```python
df_new
```

```python
# Repetindo o processo anterior para a coluna de designers
dstack = df['designer'].str.split(',', expand=True).stack().str.strip().to_frame()
df_designer = pd.get_dummies(dstack,prefix='',prefix_sep='')
df_designer = df_designer.groupby(level=0).sum()
df_new = pd.merge(df_new, df_designer, left_index=True, right_index=True)
```

```python
df_new
```

```python
# removendo o texto e fazendo o casting da coluna
df_new['time'] = df['timeOfPlay'].str.split(' ', expand=True)[0].astype('int64')
```

# Algoritmo de associção

```python
list_all_transactions = []

for mecan in df_new['mecanicas']:
    list_all_transactions += [mecan.split(',')]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# transformando uma lista desiguais em um dataframe organizado
te = TransactionEncoder()
te_ary = te.fit(list_all_transactions).transform(list_all_transactions)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df_transactions.head()
```

```python
# Itens com frequencia maior que 1%
frequent_itemsets = apriori(df_transactions, min_support = 0.01, use_colnames = True)
```

```python
# Ordenar os 30 itens mais frequentes
frequent_itemsets.sort_values(by=['support'], ascending = False).head(30)
```

```python
# top 10 itens que aparecem com mais frequencia juntos
frequent_itemsets[frequent_itemsets['itemsets'].str.len()>1].sort_values(by=['support'], ascending = False).head(10)
```

```python
# top 50 itens que aparecem com mais frequencia juntos
df_association = frequent_itemsets[frequent_itemsets['itemsets'].str.len()>1].sort_values(by=['support'])[-50:]

# Corrigindo o nome dos itens
df_association['nome'] = 0
for i in df_association.index:
    df_association['nome'][i] = ', '.join(df_association['itemsets'][i])
```

```python
# definindo a dimensão da figura
plt.figure(figsize=(8, 14))

# Criando as barras horizontais
plt.barh(y=df_association['nome'], width=df_association['support']);

# Adicionando o título
plt.title('Frequência do conjunto de mecânicas em jogos');
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules.sort_values(by=['lift'], ascending = False).drop(['antecedent support', 'consequent support', 'leverage', 'conviction'], axis=1)
```
